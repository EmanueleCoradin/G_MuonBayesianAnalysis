```r
library(ggplot2)

data_lifetime <- data_lifetime[data_lifetime>exp(5)]-exp(5)
data_precession <- data_precession[data_precession>exp(5)]-exp(5)

plotdata<-function(data, name, scale,bins=40){
  if (scale=='lin'){data=data}
  else {data=log(data)}
  histdata<-hist(data,breaks=bins,plot=FALSE)
  p <- ggplot() + theme_minimal() +
   geom_point(aes(x = histdata$mids, y = histdata$counts),color='blue',size=1) +
   geom_errorbar(aes(x=histdata$mids, ymin=histdata$counts-sqrt(histdata$counts),ymax=histdata$counts+sqrt(histdata$counts)),lwd=0.5,width=0.01, color = 'red')
  if(scale=='lin'){p<-p+labs(x = name, y = 'counts')}
  else{p<-p+labs(x =cat('log',name), y = 'counts')}
  p
}
plotdata(data_lifetime,'lifetime','lin',60)
plotdata(data_lifetime,name='lifetime','log')
plotdata(data_precession,'precession','lin',60)
plotdata(data_precession,name='precession','log')
```

```r
lifetime_hist<-hist(data_lifetime,breaks=60,plot=FALSE)
x<-lifetime_hist$mids
y<-lifetime_hist$counts

n_burnin <- 2e+3 # length of the burn-in phase
thinning <- 2
Nrep = 1e+6     # number of values to simulate

model_string <- 'model{
  # Likelihood
  for (i in 1:length(x)) {
    I[i] <- I0*exp(-x[i]/tau)+c
    y[i] ~ dpois(I[i])
  }

  # Prior
  I0 ~ dunif(0, 5000)
  c ~ dunif(0, 100)
  tau ~ dunif(0,5e+2)
}'

dataList = list(x = x, y = y)
model <- jags.model(file = textConnection(model_string), data = dataList)

update(model, n.iter = n_burnin)

posterior_sample <- coda.samples(model, variable.names = c('I0','c','tau'), n.iter = Nrep, thin = thinning)
summary(posterior_sample)
plot(posterior_sample)

posterior_matrix <- as.matrix(posterior_sample)
# Retrieve the chains
a_samples <- posterior_matrix[, "I0"]
b_samples <- posterior_matrix[, "c"]
c_samples <- posterior_matrix[, "tau"]

par(mfrow = c(2, 2))

acf(a_samples, main = "Autocorrelation of I0")
acf(b_samples, main = "Autocorrelation of c")
acf(c_samples, main = "Autocorrelation of tau")
```
